```python
import shutil  
import os
from tkinter import *
import import_ipynb
import enc_dec
from enc_dec import encrypt_file
from enc_dec import decrypt_file
 
window = Tk()
global file_name
global file_enc
global file_bck

file_name = 'C:/Users/IBM_ADMIN/SSD_Project/New folder/New folder/activity.txt'
file_enc = 'C:/Users/IBM_ADMIN/SSD_Project/New folder/New folder/activity.txt.enc'
file_bck = 'C:/Users/IBM_ADMIN/SSD_Project/New folder/New folder/activity.txt.bck'

#Window Bar Label
decrypt_file(file_enc, enc_dec.key)


def enc():
    encrypt_file(file_name, enc_dec.key)

def rem():
    os.remove(file_name)
    

def logout():
    window.destroy()

window.title("Please select from the following Options:")
 
window.geometry('500x500')
 
#Main Menu  
lbl = Label(window, text="Manage your daily activity")
 
lbl.grid(column=1, row=1)

#The first option function starts here, asks for input and updates the activities file
lbl = Label(window, text="Add a record")
 
lbl.grid(column=0, row=5)

def clicked1():
     with open(file_name, 'a+') as af:
         ll = af.read()
         print(ll)
         num = input("Enter number: ")
         act = input("Enter activity type (eg. running, walking, cycling): ")
         dis = input("Enter distance in km (eg. 5.00): ")
         tim = input("Enter time taken (eg. 0.45: ")
         pace = (float(dis) / float(tim))
         af.write('\n' + num +" " + act + " " + dis + "   " + tim + "    " + str(pace))                  
     return
     af.close() 
enc()
  
     
btn = Button(window, text="Click Me", command=clicked1)
 
btn.grid(column=1, row=5)


#This is the second option function, reads the activites file to the screen
lbl = Label(window, text="Read the contents of the file")
 
lbl.grid(column=0, row=10)

def clicked2():
    with open(file_name, 'r') as rf:
       for line in rf:
         print(rf.read())
       return
    rf.close
     
 
btn = Button(window, text="Click Me", command=clicked2)
 
btn.grid(column=1, row=10)


#This is the third option function, creates a copy of the file to .bck 
lbl = Label(window, text="create a new copy")
 
lbl.grid(column=0, row=15)
    
def clicked3():
     with open(file_name, 'r') as cf:
      shutil.copy(file_name, 'activity.txt.bck')
      encbck()
      rembck()
      return
     cf.close()
#encbck()     
#rembck()

#lbl.configure(text="File was created")
 
btn = Button(window, text="Click Me", command=clicked3)
 
btn.grid(column=1, row=15)

#This is the fourth option function, reads out the file and asks which one to delete
lbl = Label(window, text="Delete an entry")
 
lbl.grid(column=0, row=20)
    
def clicked4():
     with open(file_name, 'r+') as df:
        t = df.read()
        #for line in df:
        print(df.read())
       
        to_delete = input("Which record do you want to delete:").strip()
        df.seek(0)
        for line in t.split('\n'):
            if line != to_delete:
                df.write(line + '\n')
        df.truncate()
        return
     df.close()

btn = Button(window, text="Click Me", command=clicked4)
btn.grid(column=1, row=20)

def encbck():
    encrypt_file('activity.txt.bck', enc_dec.key)
   
def rembck():
    os.remove('activity.txt.bck')
   
lbl = Label(window, text="Please click submit to update the file")
lbl.grid(column=0, row=30)
    
btn = Button(window, text="Submit", command=rem) 
btn.grid(column=1, row=30)

lbl = Label(window, text="Please click Logout to exit the application")
lbl.grid(column=0, row=60)
    
btn = Button(window, text="Logout", command=logout) 
btn.grid(column=1, row=60)

window.mainloop()
```
